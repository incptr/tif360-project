## DQN vs 2nd winning players

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
raw_data = {'agents': ['DQN-RANDOM', 'DQN-AB', 'DQN-ONESTEP', 'DQN-NSTEP'],
        'DQN': [80, 0, 1, 0],
        '2nd_Winning_Player': [8,36,60,75],
        'Invalid_Move': [12,14,39,25]}
df = pd.DataFrame(raw_data, columns = ['agents', 'DQN', '2nd_Winning_Player', 'Invalid_Move'])
df
```

```python
# Setting the positions and width for the bars
pos = list(range(len(df['DQN']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(6,4))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        df['DQN'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='r', 
        # with label the first value in first_name
        label=df['agents'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        df['2nd_Winning_Player'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='b', 
        # with label the second value in first_name
        label=df['agents'][1]) 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['post_score'] data,
        df['Invalid_Move'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='g', 
        # with label the third value in first_name
        label=df['agents'][2]) 

# Set the y axis label
ax.set_ylabel('Games')

# Set the chart's title
ax.set_title('Performance Algorithm of DQN')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['agents'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
#plt.ylim([0, max(df['DQN'] + df['2nd_Winning_Player'] + df['Invalid_Move'])] )
plt.ylim([0, 120])

# Adding the legend and showing the plot
plt.legend(['DQN', '2nd_Winning_Player', 'Invalid_Move'], loc='upper center')
plt.grid()
#plt.show()
plt.savefig('Performance algorithm of DQN')
```
